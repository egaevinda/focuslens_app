# **Import Library**

```python
import os, glob, shutil, random, cv2
import numpy as np, pandas as pd
from sklearn.model_selection import train_test_split
```

Model deteksi wajah

```python
haar_path = cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
face_cascade = cv2.CascadeClassifier(haar_path) if CROP_FACE else None
```

KUMPULKAN FILE DARI DRIVE

```python
exts = ("*.jpg","*.jpeg","*.png","*.JPG","*.JPEG","*.PNG")
records = []
missing = []

for c in CLASSES:
    src = os.path.join(INPUT_DIR, c)
    if not os.path.isdir(src):
        missing.append(src);
        continue
    files = []
    for e in exts:
        files += glob.glob(os.path.join(src, e), recursive=False)
        files += glob.glob(os.path.join(src, "**", e), recursive=True)
    for f in sorted(set(files)):
        records.append({"path": f, "label": c})

if missing:
    print("‚ö†Ô∏è Folder yang tidak ditemukan:", *missing, sep="\n - ")

df = pd.DataFrame(records)
print("Total gambar ditemukan:", len(df))
print(df.head(3))

assert len(df) > 0, "Tidak ada gambar di DRIVE_INPUT_DIR. Pastikan ada DATASET_RAW/fokus|bosan|distraksi berisi gambar."
```

SPLIT STRATIFIKASI

```python
train_df, tmp_df = train_test_split(df, test_size=(TEST_SIZE+VAL_SIZE),
                                    stratify=df["label"], random_state=RANDOM_SEED)
rel_val = VAL_SIZE / (TEST_SIZE + VAL_SIZE)
val_df, test_df = train_test_split(tmp_df, test_size=(1-rel_val),
                                   stratify=tmp_df["label"], random_state=RANDOM_SEED)
print(f"Split ‚Üí train={len(train_df)}, val={len(val_df)}, test={len(test_df)}")
```

FUNGSI PROSES & SIMPAN


```python
def process_image(src_path, target_size=128, crop_face=False, face_model=None):
    img = cv2.imread(src_path)
    if img is None:
        return None
    if crop_face and face_model is not None:
        gray  = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = face_model.detectMultiScale(gray, 1.2, 5)
        if len(faces) > 0:
            x,y,w,h = max(faces, key=lambda f: f[2]*f[3])  # wajah terbesar
            img = img[y:y+h, x:x+w]
    img = cv2.resize(img, (target_size, target_size))
    return img

def save_split(df_split, split_name):
    rows = []
    for _, r in df_split.iterrows():
        img = process_image(r["path"], TARGET_SIZE, CROP_FACE, face_cascade)
        if img is None:
            continue
        name = os.path.splitext(os.path.basename(r["path"]))[0]
        out  = os.path.join(OUTPUT_DIR, split_name, r["label"], f"{name}.jpg")
        cv2.imwrite(out, img)
        rows.append({"path": out, "label": r["label"]})
    return pd.DataFrame(rows)

train_out = save_split(train_df, "train")
val_out   = save_split(val_df,   "val")
test_out  = save_split(test_df,  "test")
```

Simpan CSV label ke Drive


```python
train_out.to_csv(os.path.join(OUTPUT_DIR, "train_labels.csv"), index=False)
val_out.to_csv(  os.path.join(OUTPUT_DIR, "val_labels.csv"),   index=False)
test_out.to_csv( os.path.join(OUTPUT_DIR, "test_labels.csv"),  index=False)

print("‚úÖ Selesai preprocessing. Tersimpan ke:", OUTPUT_DIR)
```

#Latih model + evaluasi

```python
from google.colab import drive
drive.mount('/content/drive')

DATA_DIR = "/content/drive/MyDrive/DATASET_PROCESSED/dataset"
MODEL_DIR = "/content/drive/MyDrive/DATASET_PROCESSED/models"
REPORT_DIR = "/content/drive/MyDrive/DATASET_PROCESSED/reports"

import os; os.makedirs(MODEL_DIR, exist_ok=True); os.makedirs(REPORT_DIR, exist_ok=True)
```

Import Library

```python
import os, numpy as np, matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from sklearn.metrics import classification_report, confusion_matrix

IMG_SIZE   = 128
BATCH_SIZE = 32
EPOCHS     = 20
USE_TRANSFER = True
```

Data generators

```python
train_gen = ImageDataGenerator(
    rescale=1./255, rotation_range=10, width_shift_range=0.1,
    height_shift_range=0.1, zoom_range=0.1, horizontal_flip=True
).flow_from_directory(
    os.path.join(DATA_DIR, "train"), target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE, class_mode='categorical', shuffle=True
)

val_gen = ImageDataGenerator(rescale=1./255).flow_from_directory(
    os.path.join(DATA_DIR, "val"), target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE, class_mode='categorical', shuffle=False
)

test_gen = ImageDataGenerator(rescale=1./255).flow_from_directory(
    os.path.join(DATA_DIR, "test"), target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE, class_mode='categorical', shuffle=False
)

NUM_CLASSES = len(train_gen.class_indices)
labels = list(train_gen.class_indices.keys())
print("labels:", labels)
```

Model

```python
if USE_TRANSFER:
    base = MobileNetV2(input_shape=(IMG_SIZE, IMG_SIZE, 3), include_top=False, weights='imagenet')
    base.trainable = False  # freeze awal
    x = GlobalAveragePooling2D()(base.output)
    x = Dropout(0.3)(x)
    outputs = Dense(NUM_CLASSES, activation='softmax')(x)
    model = Model(inputs=base.input, outputs=outputs)
else:
    model = Sequential([
        tf.keras.layers.Conv2D(32,(3,3),activation='relu',input_shape=(IMG_SIZE,IMG_SIZE,3)),
        tf.keras.layers.MaxPooling2D(2,2),
        tf.keras.layers.Conv2D(64,(3,3),activation='relu'),
        tf.keras.layers.MaxPooling2D(2,2),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(128,activation='relu'),
        tf.keras.layers.Dropout(0.3),
        tf.keras.layers.Dense(NUM_CLASSES,activation='softmax')
    ])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()

ckpt = ModelCheckpoint(os.path.join(MODEL_DIR, "model_best.keras"),
                       monitor="val_accuracy", save_best_only=True, mode="max", verbose=1)
early = EarlyStopping(monitor="val_accuracy", patience=5, restore_best_weights=True)

history = model.fit(train_gen, validation_data=val_gen, epochs=EPOCHS, callbacks=[ckpt, early])

model.save(os.path.join(MODEL_DIR, "model_final.h5"))
print("‚úÖ disimpan:", os.path.join(MODEL_DIR,"model_final.h5"))
```

Plot kurva & evaluasi

```python
plt.figure(); plt.plot(history.history['accuracy']); plt.plot(history.history['val_accuracy'])
plt.legend(['train','val']); plt.title('Accuracy'); plt.xlabel('Epoch'); plt.ylabel('Acc'); plt.savefig(os.path.join(REPORT_DIR,'curve_acc.png')); plt.show()

plt.figure(); plt.plot(history.history['loss']); plt.plot(history.history['val_loss'])
plt.legend(['train','val']); plt.title('Loss'); plt.xlabel('Epoch'); plt.ylabel('Loss'); plt.savefig(os.path.join(REPORT_DIR,'curve_loss.png')); plt.show()

test_gen.reset()
pred = model.predict(test_gen, verbose=0)
y_pred = np.argmax(pred, axis=1)
y_true = test_gen.classes

print("=== Classification Report ===")
report_text = classification_report(y_true, y_pred, target_names=labels)
print(report_text)

cm = confusion_matrix(y_true, y_pred)
print("Confusion matrix:\n", cm)

# simpan report dan cm
with open(os.path.join(REPORT_DIR,'classification_report.txt'), 'w') as f:
    f.write(report_text)

import matplotlib.pyplot as plt
plt.figure(); plt.imshow(cm, interpolation='nearest'); plt.title('Confusion Matrix'); plt.colorbar()
ticks = np.arange(len(labels)); plt.xticks(ticks, labels, rotation=45); plt.yticks(ticks, labels)
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, str(cm[i,j]), ha='center', va='center')
plt.xlabel('Predicted'); plt.ylabel('True'); plt.tight_layout()
plt.savefig(os.path.join(REPORT_DIR,'confusion_matrix.png')); plt.show()

print("üìÇ Reports tersimpan di:", REPORT_DIR)
```

Fine-tuning

```python
for layer in base.layers[-30:]:
    layer.trainable = True
model.compile(optimizer=tf.keras.optimizers.Adam(1e-5),
              loss='categorical_crossentropy', metrics=['accuracy'])
history_ft = model.fit(train_gen, validation_data=val_gen, epochs=5)
model.save(os.path.join(MODEL_DIR, "model_finetuned.h5"))
```
